```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Eric-Mendes/sortify-test/main/saved_tracks_audio_features.csv")
X = df.drop("id", axis="columns")
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


pca_2d = PCA(n_components=2)
pca_3d = PCA(n_components=3)

tsne_2d = TSNE(n_components=2)
tsne_3d = TSNE(n_components=3)
```

```python
def plot_elbow(df):
  n_clusters = [i for i in range(6, 120+1)]
  n_clusters_inertia = []
  for clusters in n_clusters:
    model = KMeans(n_clusters=clusters)
    model.fit(df)
    n_clusters_inertia.append(model.inertia_)
  
  plt.plot(n_clusters, n_clusters_inertia)
  plt.plot([n_clusters[i] for i in range(0, len(n_clusters), 4)], [n_clusters_inertia[i] for i in range(0, len(n_clusters_inertia), 4)])
  plt.show()
```

```python
from sklearn.pipeline import Pipeline


pipeline1 = Pipeline(
    [
      ("scaler", scaler),
    ]
)

pipeline2 = Pipeline(
    [
      ("pca_2d", pca_2d),
      ("scaler", scaler),
    ]
)

pipeline3 = Pipeline(
    [
      ("pca_3d", pca_3d),
      ("scaler", scaler),
    ]
)

pipeline4 = Pipeline(
    [
      ("tsne_2d", tsne_2d),
      ("scaler", scaler),
    ]
)

pipeline5 = Pipeline(
    [
      ("tsne_3d", tsne_3d),
      ("scaler", scaler),
    ]
)
```

```python
plot_elbow(pipeline1.fit_transform(X))
plot_elbow(pipeline2.fit_transform(X))
plot_elbow(pipeline3.fit_transform(X))
plot_elbow(scaler.fit_transform(tsne_2d.fit_transform(X)))
plot_elbow(scaler.fit_transform(tsne_3d.fit_transform(X)))
```

```python
from sklearn.preprocessing import PolynomialFeatures


poly = PolynomialFeatures()
```

```python
plot_elbow(pipeline1.fit_transform(poly.fit_transform(X)))
plot_elbow(pipeline2.fit_transform(poly.fit_transform(X)))
plot_elbow(pipeline3.fit_transform(poly.fit_transform(X)))
plot_elbow(scaler.fit_transform(tsne_2d.fit_transform(poly.fit_transform(X))))
plot_elbow(scaler.fit_transform(tsne_3d.fit_transform(poly.fit_transform(X))))
```

```python
X_ = scaler.fit_transform(tsne_2d.fit_transform(X))
X__ = scaler.fit_transform(tsne_2d.fit_transform(poly.fit_transform(X)))

model = KMeans(
    n_clusters=20
)
model.fit_transform(X_)
print(model.inertia_)
groups1 = model.predict(X_)

model.fit_transform(X__)
print(model.inertia_)
groups2 = model.predict(X__)
```

```python
df_result = X.copy()
df_result["id"] = df["id"].copy()
df_result["groups"] = groups1

df_result.to_csv("grouped_tracks.csv", index=False)
```
